```python
# =======================================================
#  PERCEPTRN 3: VARIACIN DE CARACTERSTICAS
# =======================================================

# PASO 1锔: Importar librer铆as necesarias
import numpy as np
import pandas as pd

print("Librer铆as importadas correctamente.")
```

```python
# PASO 2锔: Cargar el dataset de diabetes desde un enlace p煤blico
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv"
df = pd.read_csv(url, header=None)

# A帽adimos los nombres de las columnas (porque el dataset no los incluye)
df.columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
              'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']

print("Dataset cargado exitosamente.")
display(df.head())
```

```python
# PASO 3锔: Seleccionar un conjunto diferente de caracter铆sticas
# Usaremos solo 4 variables para ver el impacto:
# (Menos caracter铆sticas que los modelos 1 y 2)
selected_features = ['Glucose', 'BMI', 'Age', 'Pregnancies']

X = df[selected_features].values
y = df['Outcome'].values

print("Caracter铆sticas seleccionadas:")
print(selected_features)
print(f"Total de caracter铆sticas usadas: {len(selected_features)}")
```

```python
# PASO 4锔: Normalizar los datos (es importante para que todas las variables est茅n en la misma escala)
X = (X - X.mean(axis=0)) / X.std(axis=0)
print("Datos normalizados correctamente.")
```

```python
# PASO 5锔: Dividir los datos en entrenamiento (70%) y prueba (30%)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print("Datos divididos en entrenamiento y prueba exitosamente.")
print(f"Tama帽o de entrenamiento: {X_train.shape}")
print(f"Tama帽o de prueba: {X_test.shape}")
```

```python
# PASO 6锔: Definir la funci贸n de activaci贸n escal贸n (step function)
def step_function(weighted_sum):
    """Funci贸n de activaci贸n tipo escal贸n."""
    return 1 if weighted_sum >= 0 else 0

print("Funci贸n de activaci贸n definida correctamente.")
```

```python
# PASO 7锔: Crear la clase del Perceptr贸n (implementaci贸n desde cero)
class Perceptron:
    """Perceptr贸n simple hecho desde cero."""

    def __init__(self, learning_rate=0.05, epochs=100):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Entrenamiento
        for epoch in range(self.epochs):
            total_error = 0
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = step_function(linear_output)
                error = y[idx] - y_predicted
                self.weights += self.learning_rate * error * x_i
                self.bias += self.learning_rate * error
                total_error += abs(error)
            if (epoch + 1) % 10 == 0:
                print(f"poca {epoch+1}/{self.epochs} - Errores: {total_error}")

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = [step_function(x) for x in linear_output]
        return np.array(y_predicted)

print("Clase Perceptr贸n definida correctamente.")
```

```python
# PASO 8锔: Instanciar el modelo con tasa de aprendizaje moderada y entrenarlo
perceptron_model3 = Perceptron(learning_rate=0.05, epochs=100)

# Entrenar el modelo
perceptron_model3.fit(X_train, y_train)

print("\nEntrenamiento del Perceptr贸n 3 completado con 茅xito.")
```

```python
# PASO 9锔: Realizar predicciones sobre el conjunto de prueba
y_pred = perceptron_model3.predict(X_test)

print("Predicciones realizadas correctamente.")
```

```python
# PASO : Evaluar el rendimiento del modelo
from sklearn.metrics import accuracy_score, confusion_matrix

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_error = 1 - accuracy

print("--- RESULTADOS DEL PERCEPTRN 3 ---")
print("Tipo: Variaci贸n de caracter铆sticas (menos features)")
print(f"Precisi贸n del modelo: {accuracy:.4f}")
print(f"Error de Clasificaci贸n: {classification_error:.4f}")

# Mostramos la matriz de confusi贸n en formato tabla
conf_matrix_df = pd.DataFrame(conf_matrix,
                              index=['Actual Negativo', 'Actual Positivo'],
                              columns=['Predicho Negativo', 'Predicho Positivo'])
display(conf_matrix_df)
```
